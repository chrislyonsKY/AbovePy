# Flood Risk Analysis with abovepy

Use DEM data to identify low-lying areas near rivers.

```python
import abovepy
import numpy as np
```

## Search for DEM tiles

Find DEM Phase 3 tiles covering the Frankfort area.

```python
tiles = abovepy.search(
    bbox=(-84.9, 38.15, -84.8, 38.25),
    product="dem_phase3",
    max_items=5
)
print(f"Found {len(tiles)} tiles")
tiles.head()
```

## Stream a DEM tile

```python
url = tiles.iloc[0].asset_url
data, profile = abovepy.read(url)
print(f"Shape: {data.shape}, CRS: {profile['crs']}")
```

## Compute elevation percentiles

The lowest 10% of elevations are potential flood zones.

```python
elev = data[0]
valid = elev[elev > 0]
p10 = np.percentile(valid, 10)
print(f"10th percentile: {p10:.1f} ft")
```

## Visualize flood zones

```python
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
ax1.imshow(elev, cmap='terrain')
ax1.set_title('DEM Elevation')
ax1.axis('off')

flood_mask = (elev > 0) & (elev < p10)
ax2.imshow(elev, cmap='terrain', alpha=0.5)
ax2.imshow(flood_mask, cmap='Blues', alpha=0.5)
ax2.set_title(f'Potential Flood Zones (below {p10:.0f} ft)')
ax2.axis('off')
plt.tight_layout()
plt.show()
```
