# Introduction to LiDAR Data

Access COPC point clouds from KyFromAbove.

```python
import abovepy
```

## Available LiDAR products

```python
for p in abovepy.list_products():
    if "LiDAR" in p.display_name:
        print(f"  {p.key}: {p.display_name} ({p.format})")
```

## Search for COPC tiles

```python
tiles = abovepy.search(
    bbox=(-84.87, 38.19, -84.83, 38.22),
    product="laz_phase2",
    max_items=3
)
print(f"Found {len(tiles)} COPC tiles")
tiles[["tile_id", "asset_url"]].head()
```

## Reading point clouds

Requires `pip install abovepy[lidar]`

```python
try:
    import laspy
    url = tiles.iloc[0].asset_url
    print(f"Tile URL: {url}")
except ImportError:
    print("laspy not installed. Run: pip install abovepy[lidar]")
```
