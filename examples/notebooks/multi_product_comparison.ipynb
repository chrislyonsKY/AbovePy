# Multi-Product Comparison

Compare DEM Phase 1 (5ft) vs Phase 3 (2ft) resolution.

```python
import abovepy
import numpy as np
```

## Search for tiles in the same area

```python
bbox = (-84.87, 38.19, -84.83, 38.22)
tiles_p1 = abovepy.search(bbox=bbox, product="dem_phase1", max_items=1)
tiles_p3 = abovepy.search(bbox=bbox, product="dem_phase3", max_items=1)
print(f"Phase 1: {len(tiles_p1)} tiles, Phase 3: {len(tiles_p3)} tiles")
```

## Read both tiles

```python
data_p1, _ = abovepy.read(tiles_p1.iloc[0].asset_url)
data_p3, _ = abovepy.read(tiles_p3.iloc[0].asset_url)
print(f"Phase 1: {data_p1.shape} (5ft)")
print(f"Phase 3: {data_p3.shape} (2ft)")
```

## Side-by-side comparison

```python
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
ax1.imshow(data_p1[0], cmap='terrain')
ax1.set_title('Phase 1 (5ft)')
ax1.axis('off')
ax2.imshow(data_p3[0], cmap='terrain')
ax2.set_title('Phase 3 (2ft)')
ax2.axis('off')
plt.tight_layout()
plt.show()
```
