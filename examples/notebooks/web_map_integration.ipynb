# Web Map Integration with TiTiler

Generate tile URLs for MapLibre GL JS, Leaflet, or leafmap.

```python
import abovepy
from abovepy.titiler import cog_tile_url, cog_preview_url, cog_info_url
```

## Find a tile

```python
tiles = abovepy.search(
    bbox=(-84.87, 38.19, -84.83, 38.22),
    product="dem_phase3",
    max_items=1
)
url = tiles.iloc[0].asset_url
print(f"COG URL: {url}")
```

## Generate TiTiler URLs

```python
endpoint = "http://localhost:8000"
tile_url = cog_tile_url(url, titiler_endpoint=endpoint)
preview_url = cog_preview_url(url, titiler_endpoint=endpoint)
info_url = cog_info_url(url, titiler_endpoint=endpoint)
print(f"TileJSON: {tile_url}")
print(f"Preview:  {preview_url}")
print(f"Info:     {info_url}")
```

## Display with leafmap (optional)

Requires `pip install abovepy[viz]`

```python
try:
    import leafmap
    m = leafmap.Map(center=[38.2, -84.85], zoom=13)
    m.add_tile_layer(
        url=tile_url.replace("tilejson.json", "tiles/{z}/{x}/{y}.png"),
        name="DEM Phase 3",
        attribution="KyFromAbove"
    )
    m
except ImportError:
    print("leafmap not installed. Run: pip install abovepy[viz]")
```
